```python
import pandas as pd
from MCForecastTools2 import MCSimulation2
import yfinance as yf
import hvplot.pandas
import holoviews as hv
from holoviews import opts
import numpy as np
```

```python
tickers = ['AGG', 'SPY', 'BHP', 'WES', 'WDS']
data = yf.download(tickers, period="5y")
```

```python
data = data.dropna()
custom_weights = [.2,.5,.1,.1,.1]
even_weights = [.2,.2,.2,.2,.2]  
```

```python
MC_five_year_even = MCSimulation2(
    portfolio_data = data,
    weights = [.2,.2,.2,.2,.2],
    num_simulation = 50,
    num_trading_days = 252 * 5
)
MC_custom_five_year = MCSimulation2(
    portfolio_data = data,
    weights = custom_weights,
    num_simulation = 50,
    num_trading_days = 252 * 5
)
MC_custom_five_year.calc_cumulative_return()
MC_five_year_even.calc_cumulative_return()
```

```python
custom_lower_line = hv.VLine(MC_custom_five_year.confidence_interval.iloc[0]).opts(color='red', line_width=10)
custom_upper_line = hv.VLine(MC_custom_five_year.confidence_interval.iloc[1]).opts(color='red', line_width=10)
custom_weight_hist = MC_custom_five_year.plot_distribution()
custom_weight_hist * custom_lower_line * custom_upper_line
even_lower_line = hv.VLine(MC_five_year_even.confidence_interval.iloc[0]).opts(color='red', line_width=10)
even_upper_line = hv.VLine(MC_five_year_even.confidence_interval.iloc[1]).opts(color='red', line_width=10)
even_weight_hist = MC_five_year_even.plot_distribution()
even_weight_histogram = (even_weight_hist * even_lower_line * even_upper_line).opts(xlabel='Final Cumulative Returns', ylabel='Frequency')
custom_weigh_hist = MC_custom_five_year.plot_distribution()
custom_weight_histogram = (custom_weigh_hist * custom_lower_line * custom_upper_line).opts(xlabel='Final Cumulative Returns', ylabel='Frequency')
display(even_weight_histogram.opts(title=f"Evenly Distributed Weights Across {MC_five_year_even.nSim} Simulations"))
display(custom_weight_histogram.opts(title= f"User Selected Weights Across {MC_custom_five_year.nSim} Simulations"))
```

```python
line_plot = MC_five_year_even.plot_simulation()
custom_line_Plot = MC_custom_five_year.plot_simulation()
display(line_plot.opts(title=f"{MC_five_year_even.nSim} simulations of even weight portfolio over the next {MC_five_year_even.nTrading} Trading Days." ))
display(custom_line_Plot.opts(title=f"{MC_custom_five_year.nSim} simulations of custom portfolio over the next {MC_custom_five_year.nTrading} Trading Days."))
```

```python
initial_investment = 10000
```

```python
user_choice = initial_investment
```

```python
even_table = MC_five_year_even.summarize_cumulative_return()
initial_investment = user_choice
even_ci_lower = round(even_table.loc['95% CI Lower']*initial_investment,2)
even_ci_upper = round(even_table.loc['95% CI Upper']*initial_investment,2)

print(
    f" The data simulated by our estimated forecast suggests that"
    f" there is a 95% chance that an initial investment of ${initial_investment} in your selected ASX stock portfolio"
    f" over the next 5 year period will end within the range of"
    f" ${even_ci_lower} and ${even_ci_upper}")
```
