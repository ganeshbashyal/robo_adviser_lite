```python
import pandas as pd
import hvplot.pandas
import yfinance as yf

%matplotlib inline
```

```python
# Taking user input
input1 = 'ANZ'
input2 = 'LPI'
input3 = 'BHP'
input4 = 'MHJ'
input5 = 'LOV'

ticker_list = [f'{input1}.AX', f'{input2}.AX', f'{input3}.AX', f'{input4}.AX', f'{input5}.AX']
```

```python
# Fetching input's data
x = 1
for ticker in ticker_list:
    globals()[f'input{x}_info'] = yf.Ticker(ticker).info
    x += 1

# Closing prices dataframe
close_df = yf.download(tickers, period='1y')['Close'].dropna()
```

```python
# Assigning information to variables
x = 1
while x <= 5:
    info = globals()[f'input{x}_info']
    globals()[f'input{x}_name'] = info['longName']
    globals()[f'input{x}_shortName'] = info['shortName']
    globals()[f'input{x}_sector'] = info['sector']
    globals()[f'input{x}_industry'] = info['industry']
    globals()[f'input{x}_website'] = info['website']
    globals()[f'input{x}_city'] = info['city']
    globals()[f'input{x}_state'] = info['state']
    globals()[f'input{x}_logoUrl'] = info['logo_url']

    globals()[f'input{x}_dayHigh'] = info['dayHigh']
    globals()[f'input{x}_ebitda'] = info['ebitda']
    globals()[f'input{x}_forwardPE'] = info['forwardPE']
    globals()[f'input{x}_targetLowPrice'] = info['targetLowPrice']
    globals()[f'input{x}_recommendationKey'] = info['recommendationKey']
    
    globals()[f'input{x}_profitMargins'] = info['profitMargins']
    globals()[f'input{x}_quickRatio'] = info['quickRatio']
    globals()[f'input{x}_roa'] = info['returnOnAssets']
    globals()[f'input{x}_roe'] = info['returnOnEquity']
    
    x += 1
```

```python
# Printing all data
x = 1
for ticker in ticker_list:

    # Form and display graph
    globals()[f'input{x}_plot'] = close_df.hvplot.line(y=(ticker), title=f"{globals()[f'input{x}_shortName']}'s Performance Over The Last Year", ylabel='Price ($)', legend=False)
    display(globals()[f'input{x}_plot'])

    # General information
    print(
        f"{globals()[f'input{x}_name']} ({ticker}) is a {globals()[f'input{x}_sector']} company, apart of the {globals()[f'input{x}_industry']} industry, "
        f"located in {globals()[f'input{x}_city']}, {globals()[f'input{x}_state']}. "
        f"It traded at a high of ${globals()[f'input{x}_dayHigh']} today, its forward price-to-earnings ratio is {round(globals()[f'input{x}_forwardPE'],2)} "
        f"and its EBITDA is sitting at ${globals()[f'input{x}_ebitda']}. "
    )

    # Financial information
    print(
        f"{globals()[f'input{x}_shortName']}'s profit margins are currently at {round(globals()[f'input{x}_profitMargins']*100,2)}% "
        f"and quick ratio is {globals()[f'input{x}_quickRatio']}. "
        f"The return of assets ratio is {round(globals()[f'input{x}_roa'],5)}, while the return of equities ratio is {round(globals()[f'input{x}_roe'],5)}. "
        f"According to Yahoo Finance, the target low price is ${globals()[f'input{x}_targetLowPrice']}, so it's recommended to {globals()[f'input{x}_recommendationKey']}."
    )

    # Websites
    print(globals()[f'input{x}_website'])
    print(f'https://au.finance.yahoo.com/quote/{ticker}')
    print('----------------------------------------')
    x+=1
```
